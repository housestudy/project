```python
import requests
import json
import pandas as pd
from tqdm import tqdm
```

```python
def get_gungu_info(sido_code):
    down_url = 'https://new.land.naver.com/api/regions/list?cortarNo='+sido_code
    r = requests.get(down_url,data={"sameAddressGroup":"false"},headers={
        "Accept-Encoding": "gzip",
        "Host": "new.land.naver.com",
        "Referer": "https://new.land.naver.com/complexes/102378?ms=37.5018495,127.0438028,16&a=APT&b=A1&e=RETAIL",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
    })
    r.encoding = "utf-8-sig"
    temp=json.loads(r.text)
    temp_df = pd.DataFrame(temp['regionList'])

    #temp=list(pd.DataFrame(temp['regionList'])["cortarNo"])
    return temp_df[['cortarNo','cortarName']]
```

```python
def get_dong_info(gungu_code):
    down_url = 'https://new.land.naver.com/api/regions/list?cortarNo='+gungu_code
    r = requests.get(down_url,data={"sameAddressGroup":"false"},headers={
        "Accept-Encoding": "gzip",
        "Host": "new.land.naver.com",
        "Referer": "https://new.land.naver.com/complexes/102378?ms=37.5018495,127.0438028,16&a=APT&b=A1&e=RETAIL",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
    })
    r.encoding = "utf-8-sig"
    temp=json.loads(r.text)
    temp_df = pd.DataFrame(temp['regionList'])
    #temp=list(pd.DataFrame(temp['regionList'])["cortarNo"])
    return temp_df[['cortarNo','cortarName']]
```

```python
def get_opst_list(dong_code):
    down_url = 'https://new.land.naver.com/api/regions/complexes?cortarNo='+dong_code+'&realEstateType=PRE%3AOPST&order='
    r = requests.get(down_url,data={"sameAddressGroup":"false"},headers={
        "Accept-Encoding": "gzip",
        "Host": "new.land.naver.com",
        "Referer": "https://new.land.naver.com/complexes?ms=37.478448,127.0506355,17&a=PRE:OPST&e=RETAIL",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
    })
    r.encoding = "utf-8-sig"
    temp=json.loads(r.text)
    temp_df = pd.DataFrame(temp['complexList'])

    return temp_df
```

```python
def get_opst_detail(complexNo):
    down_url = 'https://new.land.naver.com/api/complexes/overview/{0}'.format(complexNo)
    r = requests.get(down_url,data={"complexNo":complexNo},headers={
        "Accept-Encoding": "gzip",
        "Host": "new.land.naver.com",
        "Referer": "https://new.land.naver.com/complexes?ms=37.478448,127.0506355,17&a=PRE:OPST&e=RETAIL",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
    })
    r.encoding = "utf-8-sig"
    temp=json.loads(r.text)

    return temp
```

```python
total_df = pd.DataFrame()

# 서울시에 있는 구들의 Number와 이름 return
gungu_result = get_gungu_info('1100000000')
gu_cortarNo = gungu_result['cortarNo']
gu_cortarName = gungu_result['cortarName']

columns = ['complexTypeName', 'complexType', 'complexName', 'complexNo', 'totalHouseHoldCount', 'totalDongCount', 'useApproveYmd', 'minArea', 'maxArea', 'minPrice', 'maxPrice', 'minLeasePrice', 'maxLeasePrice','leasePerDealRate', 'isaleDealRestrictionCode', 'rebuildMembershipTransYn', 'livingResidenceYn', 'latitude', 'longitude']
total_df = pd.DataFrame()

for i in tqdm(range(len(gungu_result))):
    # 구들에 따른 동들의 Number와 이름 return
    dong_result = get_dong_info(gu_cortarNo[i])
    dong_cortarNo = dong_result['cortarNo']
    dong_cortarName = dong_result['cortarName']
    
    for j in tqdm(range(len(dong_result))):
        # 동에 있는 아파트들의 정보들을 가져옴
        opst_df = get_opst_list(dong_cortarNo[j])
        
        if len(opst_df) == 0:
            continue
        
        # 아파트들의 detail 정보 가져오기
        for complexNo in opst_df['complexNo']:
            opst_detail = get_opst_detail(complexNo)
            temp_dict = dict()
            
            for column in columns:
                try:
                    temp_dict[column] = opst_detail[column]
                except KeyError:
                    temp_dict[column] = None
                
            temp_opst_detail = pd.DataFrame([temp_dict])
            total_df = pd.concat([total_df,temp_opst_detail])
```

```python
total_df.to_csv('./data/naver_data_OPST_detail.csv')
```

```python
import requests
from bs4 import BeautifulSoup

def naver_real_estate_crawling():
    # 네이버 부동산 URL
    naver_real_estate_url = 'https://new.land.naver.com/complexes'

    # 웹페이지에 요청을 보내 HTML을 받아옴
    response = requests.get(naver_real_estate_url)

    # BeautifulSoup을 사용하여 HTML을 파싱
    soup = BeautifulSoup(response.text, 'html.parser')

    # 부동산 정보를 담을 리스트 초기화
    real_estate_list = []

    # 예시로 간단하게 부동산 이름만 추출하는 코드
    complex_names = soup.select('.complex_info .item_title')

    for complex_name in complex_names:
        real_estate_list.append(complex_name.text.strip())

    # 추출된 부동산 정보 출력
    for idx, complex_info in enumerate(real_estate_list, start=1):
        print(f"{idx}. {complex_info}")

if __name__ == "__main__":
    naver_real_estate_crawling()
```
