### 각 구 별로 동 가져오기

```python
def get_dong_info(gungu_code):
    down_url = 'https://new.land.naver.com/api/regions/list?cortarNo='+gungu_code
    r = requests.get(down_url,data={"sameAddressGroup":"false"},headers={
        "Accept-Encoding": "gzip",
        "Host": "new.land.naver.com",
        "Referer": "https://new.land.naver.com/complexes/102378?ms=37.5018495,127.0438028,16&a=APT&b=A1&e=RETAIL",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
    })
    r.encoding = "utf-8-sig"
    temp=json.loads(r.text)
    temp_df = pd.DataFrame(temp['regionList'])
    #temp=list(pd.DataFrame(temp['regionList'])["cortarNo"])
    return temp_df[['cortarNo','cortarName']]
```

```python
def get_apt_list(dong_code):
    down_url = 'https://new.land.naver.com/api/regions/complexes?cortarNo='+dong_code+'&realEstateType=APT&order='
    r = requests.get(down_url,data={"sameAddressGroup":"false"},headers={
        "Accept-Encoding": "gzip",
        "Host": "new.land.naver.com",
        "Referer": "https://new.land.naver.com/complexes/102378?ms=37.5018495,127.0438028,16&a=APT&b=A1&e=RETAIL",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
    })
    r.encoding = "utf-8-sig"
    temp=json.loads(r.text)
    temp_df = pd.DataFrame(temp['complexList'])
    #try:
    #   temp=list(pd.DataFrame(temp['complexList'])["complexNo"])
    #except:
    #   temp=[]
    return temp_df
```

### 2. 최종 아파트의 complextypename으로 검색했을 때, 나오는 아파트의 정보(naver_data_APT_detail.csv)

```python
def get_apt_detail(complexNo):
    down_url = 'https://new.land.naver.com/api/complexes/overview/{0}'.format(complexNo)
    r = requests.get(down_url,data={"complexNo":complexNo},headers={
        "Accept-Encoding": "gzip",
        "Host": "new.land.naver.com",
        "Referer": "https://new.land.naver.com/complexes?ms=37.4787503,127.0514187,17&a=APT:PRE&e=RETAIL",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-origin",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
    })
    r.encoding = "utf-8-sig"
    temp=json.loads(r.text)

    return temp
```

```python
total_df = pd.DataFrame()

# 서울시에 있는 구들의 Number와 이름 return
gungu_result = get_gungu_info('1100000000')
gu_cortarNo = gungu_result['cortarNo']
gu_cortarName = gungu_result['cortarName']

columns = ['complexTypeName', 'complexType', 'complexName', 'complexNo', 'totalHouseHoldCount', 'totalDongCount', 'useApproveYmd', 'minArea', 'maxArea', 'minPrice', 'maxPrice', 'minLeasePrice', 'maxLeasePrice','leasePerDealRate', 'isaleDealRestrictionCode', 'rebuildMembershipTransYn', 'livingResidenceYn', 'latitude', 'longitude']
total_df = pd.DataFrame()

for i in tqdm(range(len(gungu_result))):
    # 구들에 따른 동들의 Number와 이름 return
    dong_result = get_dong_info(gu_cortarNo[i])
    dong_cortarNo = dong_result['cortarNo']
    dong_cortarName = dong_result['cortarName']
    
    for j in tqdm(range(len(dong_result))):
        # 동에 있는 아파트들의 정보들을 가져옴
        apt_df = get_apt_list(dong_cortarNo[j])
        
        if len(apt_df) == 0:
            continue
        
        # 아파트들의 detail 정보 가져오기
        for complexNo in apt_df['complexNo']:
            apt_detail = get_apt_detail(complexNo)
            temp_dict = dict()
            
            for column in columns:
                try:
                    temp_dict[column] = apt_detail[column]
                except KeyError:
                    temp_dict[column] = None
                
            temp_apt_detail = pd.DataFrame([temp_dict])
            total_df = pd.concat([total_df,temp_apt_detail])
```

```python
total_df.to_csv('./data/naver_data_APT_detail.csv')
```
