```python
import requests
import json
import pandas as pd
from tqdm import tqdm
```

```python
total_df = pd.DataFrame()
api_key = '6a4f6575757039363834685968426e'

for i in tqdm(range(1,1273500,1000)):
    start_index = i
    end_index = i + 999
    url = 'http://openapi.seoul.go.kr:8088/{0}/json/tbLnOpendataRtmsV/{1}/{2}/'.format(api_key,start_index,end_index)
    
    response = requests.get(url)
    data = json.loads(response.text)
    data_df = pd.DataFrame(data['tbLnOpendataRtmsV']['row'])
    total_df = pd.concat([total_df,data_df])
```

```python
total_df
```

```python
total_df.to_csv('./data/Real_estate_transaction_data.csv')
```
